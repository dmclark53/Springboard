# Water Pumps: Data Wrangling
## Business Problem:
Tanzania is a developing country and access to water is very important for the health of the population. For this reason, it is vital that all water pumps are properly working. Currently, the only way to monitor pump working status is by physically visiting the site. This is time consuming and costly. Therefore, a more intelligent solution to monitor water pump status is desirable.

This project will address the following question: How can the government of Tanzania improve water pump maintenance by knowing the pump functional status in advance?

### Missing Data
Now I will handle the missing data in this dataset.

```python
missing_data = pd.concat([df_clean.isnull().sum(), 100 * df_clean.isnull().mean()], axis=1)
missing_data.columns = ['count', '% missing']
missing_data.sort_values(by='count', inplace=True)
```

```python
missing_data[missing_data['count'] > 0]
```

**scheme_name**: Almost 50% of the values are missing for this column. Since `scheme_management` includes the same information, I will drop this column.

```python
df_clean.drop(columns=['scheme_name'], inplace=True)
```

**Observations:** One curiosity I noticed was that there appeared to be roughly equal percentages of missing values between `permit` and `public_meeting` and `funder` and `installer`. 

**Question:** Are these relationships real or is there no connection?

To investigate this, I will:
1. Pick a column amongst these pairs.
2. Subset the `DataFrame` to only include non-null values for this column.
3. Count missing values in this subset.
4. Compare the count of missing values for the second column in the pair.

If the missing values count for the second column in the pair drops to roughly zero, this could indicate a relationship between missing values in this pair. If not, then there is no relationship.

I will start with the pair: `funder` and `installer`.

```python
df_has_funder = df_clean[df_clean['funder'].notnull()]
```

```python
missing_data_with_funder = pd.concat([df_has_funder.isnull().sum(), 100 * df_has_funder.isnull().mean()], axis=1)
missing_data_with_funder.columns = ['count (w/ funder)', '% missing (w/ funder)']
missing_data_with_funder.sort_values(by='count (w/ funder)', inplace=True)
```

```python
missing_data_with_funder[missing_data_with_funder['count (w/ funder)'] > 0]
```

```python
df_missing_compare = missing_data_with_funder.merge(missing_data, left_index=True, right_index=True)
```

```python
df_missing_compare.sort_values(by='count', inplace=True)
```

```python
df_missing_compare[df_missing_compare['count'] > 0]
```

**Observations:** There does seem to be a relationship between rows missing a funder and installer. Most rows missing a funder are also missing an installer. But, this is not an exact match.

Next, I will investigate the relationship between missing values for the pair: `permit` and `public_meeting`.

```python
df_has_permit = df_clean[df_clean['permit'].notnull()]
```

```python
missing_data_with_permit = pd.concat([df_has_permit.isnull().sum(), 100 * df_has_permit.isnull().mean()], axis=1)
missing_data_with_permit.columns = ['count (w/ permit)', '% missing (w/ permit)']
missing_data_with_permit.sort_values(by='count (w/ permit)', inplace=True)
```

```python
compare_missing_permit = missing_data_with_permit.merge(missing_data, left_index=True, right_index=True)
```

```python
compare_missing_permit.sort_values(by='count', inplace=True)
```

```python
compare_missing_permit[compare_missing_permit['count'] > 0]
```

**Observations:** No relationship seen between rows missing a permit value and rows missing a public meeting.

#### Handling boolean columns with missing values

```python
df_describe_object = df_clean.describe(include=np.object).T
df_boolean = df_describe_object[df_describe_object['unique'] == 2]
```

```python
df_boolean
```

**Observations:** The only two columns with boolean values are `public_meeting` and `permit`. For each of these, roughly 5% of the values are missing.

```python
boolean_columns = df_boolean.index.values
```

```python
for column in boolean_columns:
    num_true = df_clean[column].sum()
    num_false = len(df_clean[df_clean[column].notnull()]) - num_true
    print(f"Ratio of true-to-false values for {column} column is {num_true / num_false:0.3f}.")
```

Since there is a larger number of `True` values for both `permit` and `public_meeting`, I will fill in missing values with `True`.

```python
df_clean.loc[:,boolean_columns] = df_clean.loc[:,boolean_columns].fillna(value=True)
```

```python
df_clean[boolean_columns].info()
```

```python
for column in boolean_columns:
    num_true = df_clean[column].sum()
    num_false = len(df_clean) - num_true
    print(f"New ratio of true-to-false values for {column} column is {num_true / num_false:0.3f}.")
```

As can be see, by filling in missing values with `True`, the ratios between `True` and `False` does not change by much. This makes me confident in my choice for handling the missing boolean values.

```python
df_clean[boolean_columns].describe()
```

#### Additional columns
There are four additional columns with missing values. These include: `subvillage`, `funder`, `installer`, `scheme_management`.

```python
additional_missing = ['subvillage', 'funder', 'installer', 'scheme_management']
```

```python
df_clean[additional_missing].describe().T
```

Aside from `scheme_management`, they all contain lots of unique values.

Let's see the unique values for `scheme_management`:

```python
additional_missing
```

```python
for column in additional_missing:
    print(df_clean[column].value_counts().head(n=20))
    print('')
```

The columns `subvillage`, `funder`, `installer`, and `scheme_management` include the values `None`. So, I will use this value to set all missing values for these columns. Also, `funder` and `installer` include the value 0, which I will interpret as signifying empty values. Therefore, I will set all 0 values for these columns to `None`.

```python
columns_with_zeros = ['funder', 'installer']
for column in columns_with_zeros:
    df_clean.loc[df_clean[column] == '0', column] = 'None'
```

```python
df_clean.loc[:,additional_missing] = df_clean.loc[:,additional_missing].fillna(value='None')
```

```python
missing_data_clean = pd.concat([df_clean.isnull().sum(), 100 * df_clean.isnull().mean()], axis=1)
missing_data_clean.columns = ['count', '% missing']
```

```python
missing_data_clean
```

Now all missing values have been handled.

## Variable Types: Categorical vs. Non-Categorical
* Categorical variables:
    * These are variables that describe a class of things.
* Non-categorical variables:
    * These are variables that measure things.
Categorical and non-categorical variables can be decided using a data dictionary provided by the client. Driven Data provides the data dictionary for the water pump dataset [here](https://www.drivendata.org/competitions/7/pump-it-up-data-mining-the-water-table/page/25/). Using this data dictionary, I have identified the following columns at categorical variables:

```python
categorical_vars = ['funder', 'installer', 'wpt_name', 'basin', 'subvillage', 'region', 'region_code', 'district_code',
                   'lga', 'ward', 'public_meeting', 'permit', 'construction_year',
                   'extraction_type', 'extraction_type_group', 'extraction_type_class', 'management', 'management_group',
                   'payment_type', 'water_quality', 'quality_group', 'quantity_group', 'source', 'source_type', 
                    'source_class', 'waterpoint_type', 'waterpoint_type_group', 'scheme_management', 'status_group']
```

The dataset can also be reviewed for variable types by printing value counts for each column. Using visual inspection, I can decide if I have missed any categorical variables.

```python
for column in df_clean.columns:
    print(df_clean[column].value_counts().head(n=10))
    print('')
```

I missed a couple of categorical variables and will add them to the above list.

```python
categorical_vars += ['payment', 'quantity']
```

Now, I will convert all columns that are categorical variables to type `category'`.

```python
categorical_dict = {x: 'category' for x in categorical_vars}
```

```python
df_clean = df_clean.astype(categorical_dict)
```

```python
df_clean.info()
```

### Duplicate Values
First, I will investigate if there are any duplicate rows.

```python
df_clean.duplicated(keep='first').sum()
```

There are 36 rows  that are exactly duplicated. I will drop these.

```python
df_clean.drop_duplicates(keep='first', inplace=True)
```

```python
df_clean.duplicated(keep='first').sum()
```

The data type `category` will naturally have duplicate values since they designate groups of data. 

The data type `int` and `float` might have duplicate values, but since these are numeric values it is logical duplicates could occur. This said, it is worth noting the columns `longitude` and `latitude`.

```python
df_clean[['longitude', 'latitude']].value_counts().head(n=10)
```

These geographical locations come in pairs. This could indicate that data was recorded twice for each pump. Further analysis of this observation will be done during EDA.

The only additional data type to consider is `datetime64[ns]`. There is only one column that is a datetime colum and that is `date_recorded`. If we take a look at the values counts for that column we can see that there are multiple values for the same date.

```python
df_clean['date_recorded'].value_counts().head(n=10)
```

This could indicate that data for multiple pumps was recorded for each date in the dataset.

### Dropping Columns
